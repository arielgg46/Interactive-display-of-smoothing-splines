```python
# Importar librerías necesarias
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import splrep, BSpline, UnivariateSpline
%matplotlib inline

# Modo oscuro
plt.style.use('dark_background')

# Función interactiva de interpolación
# Parámetros:
# f: función a interpolar
# s1, s2: parámetros de suavizado de splrep y UnivariateSpline respectivamente
# a, b: límites del rango a analizar
# p: cantidad de puntos a interpolar
def interpolate(f, s1, s2, a, b, p):

  # puntos aleatorios (x,y) a interpolar dentro del rango [a,b]
  x = np.arange(a, b, (b-a)/p)
  e = np.random.uniform(low=-1, high=1, size=(len(x),)) # error de la función entre [-1,1]
  y =  f(x) + e

  # spline de suavizado splrep
  tck = splrep(x, y, s=s1)

  # spline de suavizado UnivariateSpline
  uni = UnivariateSpline(x, y, s=s2)

  xnew = np.arange(a, b, 1/50)

  # sombrear el área entre las curvas f(x)-1 y f(x)+1
  plt.fill_between(xnew, f(xnew)-1, f(xnew)+1, alpha=0.2)
  # graficar función original f(x)
  plt.plot(xnew, f(xnew), '-.', label='f(x)')
  # graficar spline del splrep
  plt.plot(xnew, BSpline(*tck)(xnew), '-', label=f'splrep s1={s1}')
  # graficar spline del UnivariateSpline
  plt.plot(xnew, uni(xnew), '-', label=f'UnivariateSpline s2={s2}')
  # graficar puntos de interpolación
  plt.plot(x, y, 'o')

  plt.legend()
  plt.show()


# Función a interpolar
def f(x):
  return np.sin(x)*3

# Interacción
interactive_plot = interactive(interpolate, f=fixed(f), s1=(0,100), s2=(0,100), a=fixed(0), b=fixed(8), p=fixed(30))
output = interactive_plot.children[-1]
output.layout.height = '500px'
interactive_plot
```
